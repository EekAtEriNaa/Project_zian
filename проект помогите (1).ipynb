```python
import pandas as pd
```

```python
your_df = pd.read_csv('data_all.tsv', sep='\t').drop(columns=['Unnamed: 0', 'nrooms', 'otdelka'])
```

```python
your_df.stack().str.replace(',','.').unstack()
```

```python
import numpy as np
```

```python
your_df['to_metro'] = your_df['to_metro'].astype(str)
```

```python
your_df['to_metro_transport'] = 0
```

```python
k = 0
for i in your_df['to_metro']:
    if len(i) > 15:
        your_df['to_metro_transport'][k] = i
        k+=1
    else:
        your_df['to_metro_transport'][k] = '-'
        k+=1
```

```python
your_df['to_metro_transport'].value_counts()
```

```python
l = 0
for i in your_df['to_metro']:
    if len(i) > 16:
        your_df['to_metro'][l] = '-'
        l+=1
    else:
        l+=1
```

```python
f = 0
for i in your_df['to_metro']:
    if i == 'nan':
        your_df['to_metro'][f] = '-'
        f+=1
    else:
        f+=1
```

```python
your_df['to_metro'].value_counts()
```
